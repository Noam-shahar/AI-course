```python
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt

# Generate random Celsius temperatures
celsius = np.random.uniform(low=-20, high=40, size=15)

# Convert Celsius to Fahrenheit
fahrenheit = (celsius * 9/5) + 32

# Create a simple neural network
model = tf.keras.Sequential([
    tf.keras.layers.Dense(units=1, input_shape=[1])  # Single neuron in a single layer
])

# Compile the model
model.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=0.01), loss='mean_squared_error')

# Training the model
history = model.fit(celsius, fahrenheit, epochs=500, verbose=0)

# Make a prediction for a new temperature not in the dataset
new_temp_celsius = np.array([100.0])
predicted_fahrenheit = model.predict(new_temp_celsius)
print(f"A temperature of {new_temp_celsius[0]}°C is predicted to be {predicted_fahrenheit[0][0]}°F")

# Get the learned weights of the model
weights = model.get_weights()
print(f"Learned weight (a): {weights[0][0][0]}")
print(f"Learned bias (b): {weights[1][0]}")

# Plotting MSE vs Epoch
plt.figure(figsize=(10, 6))
plt.plot(history.history['loss'])
plt.title('Mean Squared Error vs Epoch')
plt.xlabel('Epoch')
plt.ylabel('Mean Squared Error')
plt.grid(True)
plt.show()
```
